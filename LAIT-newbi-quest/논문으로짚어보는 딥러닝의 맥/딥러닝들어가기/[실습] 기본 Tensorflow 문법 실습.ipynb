# constant

```python
def print_tf(x):
  print("Type is\n %s" % (type(x)))
  print("Value is\n %s" % (x))

hello = tf.constant("Hello, It's me!")
print_tf(hello)
```

```python
a = tf.constant(1.5)
b = tf.constant(2.5)
print_tf(a)
print_tf(b)
```

# Variable
- 일반적으로 parameter

```python
weight = tf.Variable(10)
print_tf(weight)
```
